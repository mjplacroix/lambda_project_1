##Intro

Even if you don't follow Bitcoin and the world of blockchains closely, chances are you've heard of it and know its price can vary wildly. I was curious about what some of the leading predictive indicators of price were for the two leading cryptocurrencies - Bitcoin and Ethereum. 

Before we dive in, let's do a brief Blockchain 101. First, there's a difference between the network or software called Bitcoin and the currency traded on the network - which is also called Bitcoin. In most other cases, like Ethereum, this distinction is clearer. Ethereum is the network and ether is the currency. 

I started with the broad question of what factors could impact the price. I pulled several parallel datasets for each cryptocurrency, cleaned, and merged them until I had a dataset for each that included:


*   Price
*   Block Size
*   Number of Addresses
*   Block Difficulty 
*   Hash Rate
*   Number of Ether
*   Number of Transactions Per Day (Txns/day)

```
import seaborn as sns
import matplotlib.pyplot as plt
corr = btc_df.corr()
ax = sns.heatmap(
    corr, 
    center=0,
    cmap=sns.diverging_palette(0, 130, n=200),
    square=True
)
ax.set_xticklabels(
    ax.get_xticklabels(),
    rotation=45,
    horizontalalignment='right'
);

ax.set_title('Bitcoin: Price VS Other Factors')
plt.show()
```

```
corr = eth_df.corr()
ax = sns.heatmap(
    corr, 
    center=0,
    cmap=sns.diverging_palette(0, 250, n=200),
    square=True
)
ax.set_xticklabels(
    ax.get_xticklabels(),
    rotation=45,
    horizontalalignment='right'
);

ax.set_title('Ethereum: Price VS Other Factors')
plt.show()
```

## Analysis
As you can see based on this heatmap, there are 2, possibly 3 leading candidates that have a stronger correlation with price: Block Size, Hash Rate, and Txns/Day. We can immediately eliminate the Hash Rate as that is actually a lagging indicator of price. Why? Briefly, blockchain networks garner their security from many computers using their computing power to solve difficult math problems as a fancy way to demonstrate they're in agreement. Sounds odd, I know. To run all those computers requires a lot of power, power that costs a lot of money. 

A blockchain network rewards those who solve these difficult problems fastest with money. So, logically, the blockchains with the highest value will garner the largest number of computers trying to solve the problem as everyone wants the most value. The Hash Rate is a measure of how much computing power is trying to solve that problem and is a lagging indicator of price. Thus, we can eliminate it from our analysis. 

We are left with Block Size and Txns/Day as the two leading indicators.

Part of the reason that these two are both so closely correlated is that they're actually measuring fairly similar things. A major factor affecting the size of each block is the number of transactions included in that block. I'm not surprised that there is a strong corrlation between both of these and the price of Bitcoin and Ethereum because a higher (or lower) Blocksize or number of Txns/Day is partially indicative of how much it's being used, how much demand there is for the network. And, naturally, the higher (or lower) the demand for a blockchain network, the higher (or lower) the price will be. 

I find it interesting that Txns/Day has a higher correlation on Ethereum than on Bitcoin and would suggest that as an area of possible further analysis.
